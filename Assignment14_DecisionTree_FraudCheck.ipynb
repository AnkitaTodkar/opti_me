```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report
from sklearn import preprocessing
from sklearn import tree
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df_fraud= pd.read_csv("D:\Ankita\Assignments\\Fraud_check.csv")
df_fraud
```

```python
df_fraud =df_fraud.rename(columns={"Marital.Status":"marital_status","Taxable.Income":"taxable_income","City.Population":"city_population","Work.Experience":"work_experience"})
df_fraud
```

```python
df_fraud.info()
```

```python
df_fraud.describe()
```

```python
pd.isnull(df_fraud).sum()
```

```python
def split(num):
    if num <= 30000:
        return "Risky"
    else:
        return "Good"
df_fraud.taxable_income = df_fraud.taxable_income.apply(split)
```

```python
df_fraud
```

```python
df_fraud.corr()
```

```python
label_encoder = LabelEncoder()
df_fraud.Undergrad = label_encoder.fit_transform(df_fraud.Undergrad)
df_fraud.marital_status = label_encoder.fit_transform(df_fraud.marital_status)
df_fraud.Urban = label_encoder.fit_transform(df_fraud.Urban)
df_fraud
```

```python
x = df_fraud.iloc[:,df_fraud.columns!="taxable_income"]
y = df_fraud.taxable_income
x
```

```python
y
```

```python
# Splitting data into training and testing data set
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=40)
```

**Building Decision Tree Classifier using Entropy Criteria**

```python
model = DecisionTreeClassifier(criterion='entropy',max_depth=3)
model.fit(x_train,y_train)
```

```python
#PLot the decision tree
plot_tree(model)
plt.show()
```

```python
fn = ['Undergrad','marital_status','taxable_income','city_population','work_experience','Urban']
cn = ['Good','Risky']
plt.figure(figsize = (20,15))
plot_tree(model,
          feature_names = fn, 
          class_names = cn,
          filled = True)

plt.show()
```
